## 12

```python

# Calculate the number of channels for the top 10 countries
top_countries = youtube_data['Country of origin'].value_counts().head(10)

# Filter the original data for these top countries and their unemployment rates
top_countries_list = top_countries.index.tolist()
unem_rate = youtube_data[(youtube_data['Country of origin'].isin(top_countries_list)) & (youtube_data['country_rank'] == 1)][['Country of origin', 'Unemployment rate']]

# Merge the two datasets
merged_data = unem_rate.merge(top_countries.reset_index(name='Number of channels'), on='Country of origin')

# Plot the data
fig, ax1 = plt.subplots(figsize=(12, 8))

# Bar plot for the number of channels
sns.barplot(x='Country of origin', y='Number of channels', data=merged_data, ax=ax1, hue='Country of origin')
ax1.set_ylabel('Number of channels', color='b')
ax1.set_xlabel('Country of origin')

# Create a second y-axis for the unemployment rate
ax2 = ax1.twinx()
sns.lineplot(x='Country of origin', y='Unemployment rate', data=merged_data, ax=ax2, marker='o', color='r')
ax2.set_ylabel('Unemployment rate (%)', color='r')

# Set the title
plt.title('Number of YouTube Channels and Unemployment Rate by Country')

plt.show()
```

## 14.

```python
import geopandas as gpd
from shapely.geometry import Point

coordinates_count = youtube_data.groupby(['Latitude', 'Longitude']).size().reset_index(name='Count')

# Create a GeoDataFrame
geometry = [Point(xy) for xy in zip(coordinates_count['Longitude'], coordinates_count['Latitude'])]
geo_df = gpd.GeoDataFrame(coordinates_count, geometry=geometry)

# Load world map
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Plotting
fig, ax = plt.subplots(figsize=(15, 10))
world.boundary.plot(ax=ax, linewidth=1)
geo_df.plot(ax=ax, color='red', markersize=geo_df['Count'] * 10, alpha=0.6)

plt.title('YouTube Data Points on World Map')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

plt.show()
```

## 16.

```python
country_counts = youtube_data.groupby('Country of origin')['Country of origin'].value_counts().sort_values(ascending=False).head(10)
country_counts_index = country_counts.index
mean_population =youtube_data[(youtube_data['Country of origin'].isin(country_counts_index)) & (youtube_data['country_rank']==1)][['Country of origin','Population']]
```

```python
mean_population.set_index(keys='Country of origin',inplace=True)
```

```python
mean_population
```

```python
count_pop = pd.merge(country_counts,mean_population,left_index=True,right_index=True)
count_pop.reset_index()
```

```python

# Create a figure and a set of subplots
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot the first bar graph (Country counts)
color = 'tab:blue'
ax1.set_xlabel('Country of origin')
ax1.set_ylabel('Count', color=color)
count_pop.groupby('Country of origin')['count'].mean().plot(kind='bar', ax=ax1, color='darkgreen', position=0, width=0.4, alpha=0.7)
ax1.tick_params(axis='y', labelcolor=color)

# Create a twin Axes sharing the xaxis
ax2 = ax1.twinx()  
color = 'tab:red'
ax2.set_ylabel('Mean Population', color=color)  
count_pop.groupby('Country of origin')['Population'].mean().plot(kind='bar', ax=ax2, color='yellow', position=1, width=0.4, alpha=0.7)
ax2.tick_params(axis='y', labelcolor=color)

ax1.legend(['Number of channels'], loc='upper center')
ax2.legend(['Population'], loc='upper right')
```

## 19.

```python
youtube_data['years_since_creation'] = (pd.to_datetime('now') - youtube_data['created_date']).dt.total_seconds() / (365.25 * 24 * 3600)

# Calculate the average number of uploads per year
youtube_data['uploads_per_year'] = youtube_data['uploads'] / youtube_data['years_since_creation']

# Extract the year for trend analysis
youtube_data['created_year'] = youtube_data['created_date'].dt.year

# Group by year and calculate the average number of uploads per year
yearly_avg_uploads = youtube_data.groupby('created_year')['uploads_per_year'].mean().reset_index()

# Plot the average uploads per year
plt.figure(figsize=(12, 6))
plt.plot(yearly_avg_uploads['created_year'], yearly_avg_uploads['uploads_per_year'], marker='o', linestyle='-', color='r')

plt.title('Average Number of Videos Uploaded per Year Since Creation')
plt.xlabel('Year')
plt.ylabel('Average Number of Uploads per Year')
plt.grid(True)
plt.show()
```

## 20.

```python
# Calculate the number of months since creation for each channel
youtube_data['months_since_creation'] = (pd.to_datetime('now') - youtube_data['created_date']).dt.total_seconds() / (30 * 24 * 3600)

# Calculate the average number of subscribers gained per month
youtube_data['subscribers_per_month'] = youtube_data['subscribers'] / youtube_data['months_since_creation']

# Extract the month and year for trend analysis
youtube_data['created_month'] = youtube_data['created_date'].dt.month
youtube_data['created_year'] = youtube_data['created_date'].dt.year

# Group by month and calculate the average subscribers gained per month
monthly_avg_subscribers = youtube_data.groupby('created_month')['subscribers_per_month'].mean().reset_index()

# Plot the average subscribers gained per month
plt.figure(figsize=(12, 6))
plt.plot(monthly_avg_subscribers['created_month'], monthly_avg_subscribers['subscribers_per_month'], marker='o', linestyle='-', color='g')

plt.title('Average Number of Subscribers Gained per Month Since Creation')
plt.xlabel('Month')
plt.ylabel('Average Number of Subscribers Gained per Month')
plt.xticks(range(1, 13))
plt.grid(True)
plt.show()
```
